```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
train=pd.read_csv('train_cars.csv')
test=pd.read_csv('test_cars.csv')
```

```python
car=pd.concat([train,test])
```

```python
car.shape
```

```python
car.head()
```

```python
car.nunique()
```

```python
car["Brand"]=[i[0] for i in car["Name"].str.split()]
```

```python
car["M1"]=[i[1] for i in car["Name"].str.split()]
car["M2"]=[i[2] for i in car["Name"].str.split()]
car["M3"]=[i[-1] for i in car["Name"].str.split()]
```

```python
car.head()
```

```python
car.drop("Name",axis=1,inplace=True)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
car['Location'] = le.fit_transform(car['Location'])
car.sample(5)
```

```python
car["Location"].value_counts()
```

```python
def t1(i):
    if i=="Automatic":
        return 1
    else:
        return 0
car["Transmission"]=car["Transmission"].apply(t1)
```

```python
car["old"]=2022-car.Year
```

```python
car.drop("Year",axis=1,inplace=True)
```

```python
car=pd.concat([car,pd.get_dummies(car.Fuel_Type,drop_first=True)],axis=1)
```

```python
car.drop("Fuel_Type",axis=1,inplace=True)
```

```python
car.head()
```

```python
car=pd.concat([car,pd.get_dummies(car.Owner_Type,drop_first=True)],axis=1)
```

```python
car.drop("Owner_Type",axis=1,inplace=True)
```

```python
car.Mileage=pd.to_numeric(car.Mileage.replace(regex=["kmpl","km/kg"],value=""))
```

```python
car.Engine=pd.to_numeric(car.Engine.replace(regex=["CC"],value=""))
```

```python
car.Power=pd.to_numeric(car.Power.replace(regex=["bhp","null"],value=["",np.NaN]))
```

```python
car.New_Price=pd.to_numeric(car.New_Price.replace(regex=["Lakh","Cr"],value=""))
```

```python
pd.set_option("max_columns",40)
```

```python
car.head()
```

```python
car.drop(columns=["M1","M2","M3"],axis=1,inplace=True)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
car['Brand'] = le.fit_transform(car['Brand'])
car.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
car["Kilometers_Driven"]=pd.DataFrame(scaler.fit_transform(car[["Kilometers_Driven"]]))
car["Mileage"]=pd.DataFrame(scaler.fit_transform(car[["Mileage"]]))
car["Engine"]=pd.DataFrame(scaler.fit_transform(car[["Engine"]]))
car["Power"]=pd.DataFrame(scaler.fit_transform(car[["Power"]]))
```
